### Imports and Benchmark Config

```python
from spatialreasoning import BenchmarkConfig, run_full_benchmark, analyze_results

config = BenchmarkConfig(
        #models=['gemma4:e2b'],
        models=['gemma3:latest', 'gemma4:e2b', 'gemma4:latest', 'gemma4:26b', 'qwen3:latest', 'qwen3.6:27b', 'mistral:7b'],
        model_aliases={
            'gemma3:latest': 'gemma3:4b',
            'gemma4:latest': 'gemma4:e4b',
            'qwen3:latest':  'qwen3:8b',
            },
        repeats=1,
        output_directory="./results",
        filter_tags=[],  # empty = run all tasks; e.g. ['2d'] to filter by tag
        think=False
    )
```

### Run Benchmark

```python
run_full_benchmark(config)
```
